#### Defining metrics: other techniques
There are some techniques that can be used to generate new ideas, and digger deeper into a user experience than what the existing metrics might be able to get us.

To be more specific, these techniques include:
- external data
- gathering our own in-depth data (from user experience research, survey,s focus groups, etc.)
- retropective analysis (aka observational analysis)
- long-term prospective experiments
- human evaluation

More details:
3 ways to gather additional data
1. User Experience Research (UER): really deep, less # of participants
    + good for brainstorming
    + can use special equipment
    - need to validate the UER results
2. Focus groups: okay deep, okay # of participants
    + get feedback on hypotheticals
    - run the risk of group thinking (aka one or two loud voices in the group)
3. Surveys: not very deep, many # of participants
    + useful for metrics we cannot directly measure
    - can't directly compare to other metrics (since survey results can be fake)

### Metric Definition and Data Capture -- aka Build Intuition
Two steps from a high level concept of a metric to a fully realized definition:
1. to fully define what data needed to actually compute the metric (numerator, denominator; remove spams or not)
2. given the data, how to summarize the metric (average, median, etc.)

Example:

High-level metric: click-through-probability = (# of users who click) / (# of users who visit)

**Defn #1**: for each minute (or some time interval), (# of cookies that click)/ (# of cookies)
* The result might be different due to the chosen of time interval 
For example, suppose a uer/ cookie visits the site and clicks, then 5 minutes later, he visits the site but doesn't click. Then 30 seconds later, he clicks. Then 12 hours later, he visits the site again but doesn't click. 

If we choose time interval to be minute, then the # of users who click = 2, and # of users who visit = 3, so the probability = 2/3

If we choose time interval to be hour, then the # of users who click = 1, and # of users who visit = 2, so the probability = 1/2

If we choose time interval to be day, then the # of users who click = 1, and # of users who visit = 1, so the probability = 1

**Defn #2**:(# of pageviews with click within 'time interval') / (# of pageviews)

For example, a user loads the page (one pageview), then 30 seconds later, he refreshes the page (another pageview) and then clicks

For Defn #1: use the per minute interval, # of users who click = 1, # of users = 1, so the probability = 1

For Defn #2: also use the per minute interval, # of pageviews with click = 1, but # of pageviews = 2, so the probability = 1/2

**Defn #3**: (# of clicks)/ (# of pageviews) --> more like the click-through-rate rather than probability

Typically, defn #1 and #2 should be indistinguishable if we choose the same relatively short time interval. 

So we'll go with defn #2 since it's easier to compute.

### Summary Metrics
We want to summarize single events to a summarized metric. 

To choose the summarized metric (eg, mean, median, 20th percentile, 90th percentile), we need to establish characteristics for the metric. 

The first one is **sensitivity and robustness**. 

The second one to consider is the **distribution** of the metric. 

Four categories of summarized metrics to keep in mind:
1. sums and counts
2. means, medians, 25th, 75th and 90th percentiles
3. probabilities and rates
4. ratios

#### Common distributions in online data
Lets talk about some common distributions that come up when you look at real user data.

For example, lets measure the rate at which users click on a result on our search page, analogously, we could measure the average staytime on the results page before traveling to a result. In this case, youd probably see what we call a **Poisson distribution**, or that the stay times would be **exponentially distributed**.

Another common distribution of user data is a power-law, **Zipfian or Pareto distribution**. That basically means that the probability of a more extreme value, z, decreases like 1/z (or 1/z^exponent). This distribution also comes up in other rare events such as the frequency of words in a text (the most common word is really really common compared to the next word on the list). These types of **heavy-tailed distributions** are common in internet data.

Finally, you may have data that is a **composition of different distributions** - latency often has this characteristic because users on fast internet connection form one group and users on dial-up or cell phone networks form another. Even on mobile phones you may have differences between carriers, or newer cell phones vs. older text-based displays. This forms what is called a mixture distribution that can be hard to detect or characterize well.

The key here is not to necessarily come up with a distribution to match if the answer isnt clear - that can be helpful - but to choose summary statistics that make the most sense for what you do have. If you have a distribution that is lopsided with a very long tail, choosing the mean probably doesnt work for you very well - and in the case of something like the Pareto, the mean may be infinite!

Four categories of summarized metrics to keep in mind:
1. sums and counts (eg., # of users who visited page)
2. means, medians, 25th, 75th and 90th percentiles (eg., mean age of users who completed a course; or median latency of page load) -- we can choose between these different options by looking at the distribution of the data
3. probabilities and rates (probability has 0 or 1 outcome in each case; rate has 0 or more)
4. ratios (eg., Prob(revenue-generating click)/ P(any click)

#### Sensitivity and Robustness
The metric should be sensitive, aka move due to the changes we care about, and robust, aka doesn't move a lot when nothing interesting has happened. 
- mean is sensitive to outliers (aka mean is not that robust)

Two main ways to measure sensitivity and robustness:
1. running experiments
2. retrospective analysis of logs

Measuring sensitivity and robustness -- choose summary metric for latency of a video

Metrics to choose from: median, 80th percentile, 85th percentile, 90th percentile, and 95th percentile

Method I: checking past experiments -- for example, we make a change on the resolution and we would expect a change in latency. So we looked into past experiments, and we notice that a better resolution typically leads to longer latency, which is expected. While comparing distributions of different metrics, we notice that median and 80th percentiles are not moving between different video resolutions, so we know that these two metrics are not sensitive enough -- they are not showing a change when we do make a change that we care about.

Method II: retrospective analysis -- for example, check the distribution of latency of different comparable videos, then draw plots of summary metrics between them to see if there are any zigzags. There are zigzags for 90th and 95th percentiles, so it means that these two metrics aren't robust enough, or the videos are not that comparable (which isn't the case here).

In conclustion, we should choose the 85th percentile, which is both sensitive and robust enough.

### How to compute the comparison between experiment and control --> Absolute vs. relative difference
Suppose you run an experiment where you measure the number of visits to your homepage, and you measure 5000 visits in the control and 7000 in the experiment. Then the **absolute difference** is the result of subtracting one from the other, that is, 2000. The **relative difference** is the absolute difference divided by the control metric, that is, 40%.

#### Relative differences in probabilities
For probability metrics, people often use **percentage points** to refer to absolute differences and **percentages** to refer to relative differences. For example, if your control click-through-probability were 5%, and your experiment click-through-probability were 7%, the absolute difference would be 2 percentage points, and the relative difference would be 40 percent. However, sometimes people will refer to the absolute difference as a 2 percent change, so if someone gives you a percentage, it's important to clarify whether they mean a relative or absolute difference!

### Characterize the Variability
#### Calculating variability
To calculate a confidence interval, we would need:
* variance (or standard deviation)
* distribution

For example, previously, we made an assumption of binomial distribution:
SE = sqrt(p_hat * (1 - p_hat) / N);
margin of error m = z* * SE

| type of metric | distribution (given large sample) | estimated variance |
|:--------:|:--------:|:--------:|
|  probability   |  binomial (normal)   |  p_hat * (1 - p_hat) / N   |
|  mean   |  normal   |  sigma_2 / N   |
|  median/ percentile   |  depends   |  depends   |
|  count/ difference   |  normal (maybe)  |  var(X) + var(Y)   |
|  rates   |  poisson   |  X_bar   |
